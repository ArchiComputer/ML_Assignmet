```python
from google_play_scraper import Sort, reviews_all
import pandas as pd

# Fetch reviews for the app 'com.dki.spb_android'
result_hanacap = reviews_all(
    'com.dki.spb_android',
    sleep_milliseconds=0,
    lang='ko',
    country='kr',
    sort=Sort.MOST_RELEVANT
)

# Convert the reviews data to a DataFrame
df_reviews = pd.DataFrame(result_hanacap)

# Display the DataFrame (optional)
print(df_reviews)
```
